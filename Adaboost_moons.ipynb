```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X, y = make_moons()
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=123)
```

The following code trains an AdaBoost classifier based on 200 Decision Stumps using
Scikit-Learn’s AdaBoostClassifier class (as you might expect, there is also an Ada
BoostRegressor class). A Decision Stump is a Decision Tree with max_depth=1—in
other words, a tree composed of a single decision node plus two leaf nodes. This is
the default base estimator for the AdaBoostClassifier class

```python
ada_clf1 = AdaBoostClassifier(
    base_estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=200,
    algorithm="SAMME.R",
    learning_rate=0.4
)


ada_clf1.fit(X_train, y_train)
```

```python
pred1 = ada_clf1.predict(X_test)
print("Accuracy of adaboost with default single stump dtree setting:",accuracy_score(y_test, pred1))
```

```python
ada_clf2 = AdaBoostClassifier(
    base_estimator=SVC(),
    n_estimators=200,
    algorithm="SAMME",
    learning_rate=0.25
)


ada_clf2.fit(X_train, y_train)
```

```python
pred2 = ada_clf2.predict(X_test)
print("Accuracy of adaboost with svc classifier with radial base kernel is:", accuracy_score(y_test, pred2))
```
